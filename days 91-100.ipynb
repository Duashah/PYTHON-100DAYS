```python
#Day 91 generators example
def my_generator():
    for i in range(500):
    
        yield i

# generator object
gen = my_generator()


# Using generator in a loop
for j in gen:
    
    print(f"Generated value: {j}")
```

DAY 92 FUNCTION CACHING

```python
#caching is use to temprorily store the results of functions so when called with same arguments then the results will be present quickly

from functools import lru_cache
import time

# Define function with caching
@lru_cache(maxsize=None)
def fx(n):
    # time-consuming computation
    time.sleep(5)
    return n * 5

# Calling the function with different arguments and printing the results
print("Calling fx(20)...")
print(fx(20))  # sleep for 5 seconds and then return 
print("done for 20")

print("Calling fx(2)...")
print(fx(2))   
print("done for 2")

print("Calling fx(6)...")
print(fx(6))  
print("done for 6")

# This calls will return quickly because the results are cached
print("Calling fx(20) again...")
print(fx(20))  # This will return immediately 
print("done for 20")

print("Calling fx(2) again...")
print(fx(2)) 
print("done for 2")

print("Calling fx(6) again...")
print(fx(6))  
print("done for 6")

print("Calling fx(61)...")
print(fx(61))  
print("done for 61")
```

DAY 94 EXERCISE 11 SOLUTION

```python
#Exercise 11 solution for windows
#Reminder for drinking water

import time
from plyer import notification

def send_notification():
    notification.notify(
        title='Reminder',
        message='Time to drink some water!',
        app_name='Water Reminder',
        timeout=10
    )

def main():
    while True:
        send_notification()
        # Wait for 1 hour (3600 seconds)
        time.sleep(3600)

if __name__ == "__main__":
    main()

#working properly, successful
```

```python
import asyncio
# without nest_asyncio

async def say_hello():
    print("Hello")
    await asyncio.sleep(5)  #  delay
    print("World")

async def main():
    # Schedule say_hello to run
    task = asyncio.create_task(say_hello())

    print("Starting task...")
    await task  # Wait for the task to complete
    print("Task finished!")

try:
    # event loop
    loop = asyncio.get_event_loop()
    if loop.is_running():
        # Run the main in existing event loop
        loop.create_task(main())
    else:
        # Run the main in event loop
        loop.run_until_complete(main())
except RuntimeError:
    # In case there's no running event loop
    asyncio.run(main())
```

DAY 97 MULTITHREADING 

```python
#Day97
#multithreading, means running different programs parallel


import threading
import time
from concurrent.futures import ThreadPoolExecutor

# Indicates some task being done
def func(seconds):
  print(f"Sleeping for {seconds} seconds")
  time.sleep(seconds)
  return seconds

def main():
  time1 = time.perf_counter()

  
  # Same code using Threads
  t1 = threading.Thread(target=func, args=[4])
  t2 = threading.Thread(target=func, args=[2])
  t3 = threading.Thread(target=func, args=[1])
  t1.start()
  t2.start()
  t3.start()
  
  t1.join()
  t2.join()
  t3.join()
  # Calculating Time 
  time2 = time.perf_counter()
  print(time2 - time1)


def poolingDemo():
  with ThreadPoolExecutor() as executor:
   
    l = [3, 5, 1, 2]
    results = executor.map(func, l)
    for result in results:
      print(result)


poolingDemo()
```

DAY 98 MULTIPROCESSING

```python
import concurrent.futures
import requests
import os

def downloadFile(url, name):
    try:
        print(f"Started Downloading {name}")
        response = requests.get(url)
        response.raise_for_status()  # Checking if request was successful
        os.makedirs('filesday98', exist_ok=True) 
        file_path = f"filesday98/file{name}.jpg"
        with open(file_path, "wb") as file:
            file.write(response.content)
        print(f"Finished Downloading {name}")
        return f"Downloaded {name}"
    except Exception as e:
        print(f"Error downloading {name}: {e}")
        return f"Failed {name}"

url = "https://picsum.photos/200/300"  # To download random photos

with concurrent.futures.ThreadPoolExecutor() as executor:  # Use ThreadPoolExecutor for I/O-bound tasks
    l1 = [url for _ in range(10)]
    l2 = [i for i in range(10)]
    results = executor.map(downloadFile, l1, l2)
    for result in results:
        print(result)
```

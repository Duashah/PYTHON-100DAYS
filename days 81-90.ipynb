DAY 87 SHUTIL MODULE

```python
#day 87

import shutil
import os

# paths
source_file = 'source_fileday87.txt'
destination_file = 'destination_file.txt'
source_directory = 'example_for_day87'
destination_directory = 'destination_directory'


# 1. Create a sample file
os.makedirs(source_directory, exist_ok=True)
with open(source_file, 'w') as f:
    f.write('This is a sample file.')

with open(os.path.join(source_directory, 'sampleday87.txt'), 'w') as f:
    f.write('This is a sample file in the source directory.')

# 2. Copy a file
shutil.copy(source_file, destination_file)
print(f'Copied {source_file} to {destination_file}')

# 3. Move a file
shutil.move(source_file, 'moved_fileday87.txt')
print(f'Moved {source_file} to moved_file.txt')

# 4. Copy a directory
shutil.copytree(source_directory, destination_directory)
print(f'Copied directory {source_directory} to {destination_directory}')

# 5. Remove a directory
shutil.rmtree(destination_directory)
print(f'Removed directory {destination_directory}')


```

```python
#by request module we can download anything from online
# using wikipedia website, article about list of cities in pakistan
import requests
from bs4 import BeautifulSoup

# URL of the website to scrape
url = "https://en.wikipedia.org/wiki/List_of_cities_in_Pakistan_by_population"


response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')


print(soup.prettify())

# Find all <h1> tags 
for article_title in soup.find_all("h1"):
    print(article_title.text)
, we can 
#this will download the html website, and by more scraping we can easily get the data we can do many thing
```
